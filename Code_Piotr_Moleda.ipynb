```python
import pandas as pd

import numpy as np

from pandas import read_csv

from statistics import mean, median, mode, stdev, pstdev, variance, pvariance

import matplotlib.pyplot as plt

import scipy.stats as stats

# Question 1A

print('---------------------------------------------------------------------------------------------------')
print('Question 1A - Compute central and dispersion measures for the attributes (math score, reading score)')
print('---------------------------------------------------------------------------------------------------')
exams=pd.read_csv('exams.csv')

exams=exams.dropna()

c1=exams['math score']
c2=exams['reading score']

print(' ')
print(' ')
print('Math Score: ')
print(' ')
print('Mean -->', (mean(c1)))
print('Median -->', (median(c1)))
print('Mode -->', (mode(c1)))
print('Standard deviation -->', (stdev(c1)))
print('Population standard deviation -->', (pstdev(c1)))
print('Variance -->', (variance(c1)))
print('Population Variance -->', (pvariance(c1)))
print('---------------------------------------------------------------------------------------------------')
print('Reading Score:')
print(' ')
print('Mean -->', (mean(c2)))
print('Median -->', (median(c2)))
print('Mode -->', (mode(c2)))
print('Standard deviation -->', (stdev(c2)))
print('Population standard deviation -->', (pstdev(c2)))
print('Variance -->', (variance(c2)))
print('Population Variance -->', (pvariance(c2)))
print(' ')

# Question 1B

print(' ')
print('---------------------------------------------------------------------------------------------------')
print('Question 1B - Describe the data attribute “race/ethnicity” using the appropriate graph type')
print('---------------------------------------------------------------------------------------------------')

exams = pd.read_csv('exams.csv')

race_type = exams['race/ethnicity'].value_counts()

plt.pie(race_type.values, labels=race_type.index, autopct='%1.1f%%', startangle=90)

plt.title('Distribiution of Race / Ethnicity')

plt.axis('equal')

plt.show()

print(' ')

# Question 1C

print(' ')
print('------------------------------------------------------------------------------------------------------------')
print('Question 1C - For attributes (writing score), use the three-sigma interval to specify the outliers if any.')
print('------------------------------------------------------------------------------------------------------------')

exams = pd.read_csv('exams.csv')

writing_mean=exams['writing score'].mean()
writing_std=exams['writing score'].std()

lower=writing_mean - 3 * writing_std
upper=writing_mean + 3 * writing_std

outliers=exams[(exams['writing score'] < lower) | (exams['writing score'] > upper)]

if outliers.empty:
  print(' ')
  print('Didn’t find outliers')
  print(' ')
else:
  print(' ')
  print(outliers)
  print(' ')

# Question 2A

print('------------------------------------------------------------------------------------------------------------')
print('Quiestion 2A - Suppose the owner of a bakery knows that the mean daily demand for his wholemeal bread is 400 loaves and the standard deviation is 20. What is the probability that the demand for its bread will be up to 450 loaves?')
print('------------------------------------------------------------------------------------------------------------')

# Values

mean=400

std_dev=20

demand=450

z=(demand - mean) / std_dev

probability_bread = stats.norm.cdf(z)
print(' ')
print('Probability: ', probability_bread)
print(' ')

# Question 2B

print(' ')
print('------------------------------------------------------------------------------------------------------------')
print('Quiestion 2B - In a school of 320 students, 85 students are in the band, 120 students are on sports teams, and 60 students participate in both activities. If a student is selected randomly, what is the probability that the student will participate in any of the activities?')
print('------------------------------------------------------------------------------------------------------------')

# Values

total_studens = 320

band_students = 85

sport_students = 120

both_activities = 60

probability_activities = (band_students / total_studens) + (sport_students / total_studens) - (both_activities / total_studens)

print(' ')
print('Probability: ', probability_activities)
```
